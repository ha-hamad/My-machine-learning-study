```python
import numpy as np
y = df['price'].values
X = df['sqft_living'].values[:,np.newaxis]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=30)
```

# Step 4

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model.fit(X_train,y_train)
predicted=model.predict(X_test)
```

# Step 5

```python
plt.scatter(X_test,y_test,color="r")
plt.title("Linear Regression")
plt.ylabel("Price")
plt.xlabel("sqft_living")
plt.plot(X_test,model.predict(X_test),color="k")
plt.show()
```

```python
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

```python
print("MSE", mean_squared_error(y_test,predicted))
print("R squared", metrics.r2_score(y_test,predicted)*100)
```

As we can see, Linear Regression doesn't seem to be a good model as we got a 46% when testing the prediction.
